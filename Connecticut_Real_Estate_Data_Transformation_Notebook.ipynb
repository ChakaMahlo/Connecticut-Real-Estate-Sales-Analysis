#### The dataset has 1054159 Rows and 14 Columns, with 2 columns of integer type, 3 columns of float type and 9 columns of object type 

#### Checking the number of missing values/Empty Spaces/NULL values in each column

```python
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print (missing_data[column].value_counts())
    print("") 
```

##### True represents NULL values/Empty spaces while 'False' represents valid values
#### The last four columns 'Non Use Code','Assessor Remarks','OPM Remarks' and 'Location' have higher percentage of NULL values,above 50% while 'Address' contains 51 and 'Date Recorded'  with 2 NULL values 

##### Non Use Code with 71.3%, Assessor Remarks: 84.7%, OPM Remarks: 98.9% and Location: 75.8%

###### The solution is to discard the three; 'Non Use Code', 'OPM Remarks' and 'Location' columns from the dataset

#### Removing the three columns: 'Non Use Code', 'OPM Remarks' and 'Location'

```python
df.drop('Non Use Code', axis =1 , inplace=True)
df.drop('OPM remarks', axis =1 , inplace=True)
df.drop('Location', axis =1 , inplace=True)
df.head(3)
```

#### 'Date Recorded' Column has two NULL values, showing the rows that corresponds to those two values

```python
df[df['Date Recorded'].isnull()]
```

#### The two rows contains unuseful or redundant data hence we delete them

```python
df = df.dropna(subset = ['Date Recorded'])
df[df['Date Recorded'].isnull()]
```

##### Rows deleted

#### The address column now has 49 blank spaces / NULL values.Displaying them

```python
df[df['Address'].isnull()]
```

##### The observation is,if the Assessed value and the Sales Amount are both 0,the data is redundant or unimportant or has no value hence the solution is to remove the rows.Showing the rows where 'Assessed Value' = 0 and 'Sale Amount' = 0

```python
df[(df['Assessed Value'] == 0) & (df['Sale Amount'] == 0) ]
```

#### Deleting the rows where both The Assessed Value and Sales Amount values is 0

```python
condition = (df['Assessed Value'] == 0) & (df['Sale Amount'] == 0) 
df = df[~condition]
df[df['Address'].isnull()]
```

#### Replacing Null values of the 'Address' Column

```python
df['Address'].fillna('Unknown',inplace=True)
```

```python
missing_data = df['Address'].isnull()
print (missing_data.value_counts())
print("")
```

### The 'Address' column now has no NULL values

#### Replacing the NULL values of the Property Type and Residential Type

```python
df['Property Type'].fillna('Unknown',inplace=True)
df['Residential Type'].fillna('Non Residential',inplace=True)
df['Assessor Remarks'].fillna('No Comment',inplace=True)
df.head()
```

#### Checking for all columns missing values

```python
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print (missing_data[column].value_counts())
    print("")
```

#### All the missing values and in all the columns are handled,no column contains the missing values

#### Handling the columns data types,first displaying each column's data type

```python
df.dtypes
```

##### Converting 'Date Recorded' column to type Date/Time,Town,Address,Property Type,Residential Type and Assessor Remarks to (str),Assessed Value and Sale Amount to (int)

```python
df['Town'] = df['Town'].astype(str)
df['Address'] = df['Address'].astype(str)
df['Assessed Value'] = df['Assessed Value'].astype(int)
df['Sale Amount'] = df['Sale Amount'].astype(int)
df['Property Type'] = df['Property Type'].astype(str)
df['Residential Type'] = df['Residential Type'].astype(str)
df['Date Recorded'] = pd.to_datetime(df['Date Recorded'])
df['Assessor Remarks'] = df['Assessor Remarks'].astype(str)
print(df.dtypes)
```

```python
df.head()
```

#### 'Sales Ratio' column can be a calculated column hence for memory purposes,we delete it

```python
df.drop('Sales Ratio', axis =1 , inplace=True)
df.head()
```
